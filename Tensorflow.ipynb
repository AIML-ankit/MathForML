```python
tf.random.shuffle(value=unsuffled_variable,seed =42)
```

So in above two example seed value is not working. 

```python
tf.random.set_seed(42)
tf.random.shuffle(value=unsuffled_variable)
```

```python
tf.random.set_seed(42)
tf.random.shuffle(value=unsuffled_variable)
```

Here if we giving tf.random.set.seed value than seed is working, why not if we giving seed value inside randon.shuffle? lets research this.

1 - If both the global and the operation seed are set: Both seeds are used in conjunction to determine the random sequence.

```python
tf.random.set_seed(42)  #Global seed
tf.random.shuffle(value=unsuffled_variable,seed =31) #operation seed
```

2- If the operation seed is set, but the global seed is not set: A default global seed and the specified operation seed are used to determine the random sequence. means no effect of operation seed.

```python
tf.random.shuffle(value=unsuffled_variable,seed =31)
```

3- If the global seed is set, but the operation seed is not: The system deterministically picks an operation seed in conjunction with the global seed so that it gets a unique random sequence.

```python
tf.random.set_seed(42)
tf.random.shuffle(value=unsuffled_variable)
```

# Creating tensor from numpy array.
Why we need this?

Calculation is much fast using numpy array as compare to tensorflow tensor

```python
numpy_A = np.arange(1,25,dtype=np.int16)
numpy_A
```

```python
tf_numpy_A = tf.constant(numpy_A)
tf_numpy_A
```

Here we can change the shape of numpy array by shpe option

```python
tf_numpy_B = tf.constant(numpy_A,shape=(2,3,4))
tf_numpy_B
```

```python
tf_numpy_C = tf.constant(numpy_A,shape=(4,3,2),dtype=tf.int32)
tf_numpy_C
```

Here we can also change the dtype

```python
tf_numpy_D = tf.constant(numpy_A,shape=(2,3,4),dtype=tf.int32)
tf_numpy_D
```
